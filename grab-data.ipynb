```python
data = data.drop(["Timestamp", "Out of all of the genres below, what are your top 3 favorites?", "Unnamed: 17"], axis=1)
output = pd.DataFrame(columns=["movie_name", "rating"])
for col in data.columns:
    output = output.append({"movie_name": col, "rating": data[col].iloc[0].split(" - ")[0]}, ignore_index=True)
data = pd.read_csv("/kaggle/input/user-data/Movielens.csv")
genres = pd.DataFrame(columns=["genre"])
genres_list = []
for genre in data["Out of all of the genres below, what are your top 3 favorites?"].iloc[0].split(";"):
    genres_list.append(genre)
genres["genre"] = genres_list
```

```python
output.to_csv("movie_ratings.csv")
genres.to_csv("movie_genres.csv")
```
